```python
import requests
import pandas as pd
import numpy as np 

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 10)
```

```python
r = requests.get('https://strazha-data2.eu-central-1.linodeobjects.com/parliaments/47/index.json')
```

```python
data = r.json()

pgMemberships = pd.DataFrame(data['pgMemberships'])
parlGroups = pd.DataFrame(data['parlGroups'])
persons = pd.DataFrame(data['persons'])
```

```python
persons = persons.merge(pgMemberships, how='left', left_on='_id', right_on='person')
persons = persons.merge(parlGroups, how='left', left_on='parlGroup', right_on='_id')

persons
```

```python
persons = persons.drop(['firstNameMistakenAs', 'lastNameMistakenAs', 'email', 'photos', 'govIds', 
              'facebook', 'website', 'about', 'maritalStatus', 'placeOfBirth',
              '_id_x', '_id_y', 'started_x', 'started_y', 'slug_x', 'slug_y', '__v_x', '__v_y',
              'started_x', 'started_y', 'ended_x', 'ended_y', 'parliament_x', 'parliament_y', '__v'], axis=1)
```

```python
df = pd.DataFrame()

for i in range(7):
    r = requests.get(f'https://strazha-data2.eu-central-1.linodeobjects.com/sessions/61aa046720e632bbcca6e9ce-2022-{i}.json')

    session_ids = list(map(lambda x: x['slug'], r.json()['parlSessions']))

    for session_id in session_ids:
        r = requests.get(f'https://strazha-data2.eu-central-1.linodeobjects.com/sessions/{session_id}/index.json')
        print(session_id)

        statement_ids = np.arange(r.json()['statementCount'] / 5, dtype='int')

        for statement_id in statement_ids:
            r = requests.get(f'https://strazha-data2.eu-central-1.linodeobjects.com/sessions/{session_id}/steno/{statement_id}.json')
            
            print(statement_id)
            
            sub_df = pd.DataFrame(r.json()['sessionStatements'])
            sub_df['date'] = session_id
            
            df = df.append(sub_df)

        print('-----')
```

```python
parl_members = parl_members.merge(persons, how='left', left_on='person', right_on='person')
```

```python
import nltk
import re
from nltk.tokenize import word_tokenize
from stop_words import get_stop_words

stop_words = get_stop_words('bg')
stop_words = stop_words + ['уважаеми', 'уважаема', 'уважаем', 'госпожа', 'госпожица', 'господин', 'дами', 'господа', 'председател', 'председателка', 'председатели', 'представителка', 'представител', 'представители', 'колеги', 'колежка', 'колега']
long_string = ' '.join(list(parl_members['paragraphs'].values))
long_string = re.sub(r'[^\w\s]', '', long_string)

word_tokens = word_tokenize(long_string)

long_string = ' '.join([w for w in word_tokenize(long_string) if w.lower() not in stop_words])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
cv = CountVectorizer(max_df=0.5, min_df=2, stop_words=stop_words)
dtm = cv.fit_transform(parl_members.paragraphs)
```

```python
LDA = LatentDirichletAllocation(n_components=10, random_state=42)
```

```python
LDA.fit(dtm)
```

```python
for i, topic in enumerate(LDA.components_):
    print([cv.get_feature_names()[index] for index in topic.argsort()[-25:]])
    print()
```

```python
parl_members['topic'] = LDA.transform(dtm).argmax(axis=1)
```
